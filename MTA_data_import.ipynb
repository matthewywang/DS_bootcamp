```python
list_final = [x.strftime('%y%m%d') for x in list_all]
```

```python
list3 = []
for i in list_final:
    url = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_" + i + ".txt"
    df = pd.read_csv(url)
    list3.append(df)
```
