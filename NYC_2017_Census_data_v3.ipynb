### NYC Subway Station Data

```python
import pandas as pd
```

```python
nyc_df = pd.read_csv('http://web.mta.info/developers/data/nyct/subway/Stations.csv')
```

```python
nyc_df.info()
```

```python
nyc_df.head(10)
```

```python
nyc_df[nyc_df["Stop Name"] == "86 St"]
```

```python
nyc_df2 = nyc_df.groupby(['Stop Name','Daytime Routes'])[['Structure']].count().sort_values(by='Structure',ascending=False)
```

```python
nyc_df2['Structure'].unique()
```

```python
nyc_df2.reset_index(inplace=True)
```

```python
nyc_df2[nyc_df2['Stop Name'] == '86 St']
```

```python
nyc_df2.groupby('Stop Name')[['Structure']].sum().sort_values(by='Structure',ascending=False).head(20)
```

```python
nyc_df.loc[nyc_df['Stop Name'].str.contains('Roose')]
```

```python
nyc_df.loc[nyc_df['Division'] == 'RIT']
```

```python
# note there are two Broadway stations, one in Queens, one in Brooklyn
nyc_df.loc[nyc_df['Stop Name'] == 'Broadway']
```

```python
list3 = list(nyc_df['Borough'].unique())
list3
```

```python
list4 = ['Queens', 'Manhattan', 'Brooklyn', 'Bronx', 'StatenIsland']
```

```python
dict1 = dict()
```

```python
for i in range(len(list3)):
    key = list3[i]
    dict1[key] = list4[i]
```

```python
dict1
```

```python
nyc_df['Borough2'] = nyc_df.Borough.map(dict1)
```

```python
nyc_df.head()
```

```python
nyc_df['Station_Name'] = nyc_df['Stop Name'] + "_" + nyc_df['Borough2']
```

```python
nyc_df.head(10)
```

```python
nyc_df.info()
```

```python
len(nyc_df['Station_Name'].unique())
```

```python
from collections import defaultdict
```

```python
dict2 = defaultdict(int)
```

```python
for x in list(nyc_df['Station_Name']):
    dict2[x] += 1
```

```python
dict2
```

```python
nyc_df.loc[nyc_df['Station_Name'] == 'Canal St_Manhattan']
```

```python
# duplicates but lat/long are approx the same, Division is different, Daytime Routes are different
nyc_df.loc[nyc_df['Station_Name'] == 'Times Sq - 42 St_Manhattan']
```

```python
nyc_df['Station_Division'] = nyc_df['Stop Name'] + "_" + nyc_df['Division']
```

```python
nyc_df.head()
```

```python
nyc_subway = nyc_df.groupby(['Division','Stop Name','Borough','Station_Division','Borough2'])[['GTFS Latitude', 'GTFS Longitude']].mean()
nyc_subway.reset_index(inplace=True)
```

```python
nyc_subway.head(10)
```

```python
nyc_subway.info()
```
